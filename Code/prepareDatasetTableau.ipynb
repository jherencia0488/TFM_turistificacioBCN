En aquest notebook, calcularem les variables que ens interessen per la visualització a Tableau, i de la mateixa manera, crearem una columna amb el nom Indicador els valors dels quals seràn l'atribut d'indicador per mostrar, i una altre columna amb el valor a mostrar, per tal de no tenir cada variable en una columna diferent.

```python
import pandas as pd

#Carreguem l'arxiu
df = pd.read_excel('../Data_cleaning/def_dataset/gent_vars_bcn.xlsx')
print(df.head())
```

```python
#Extraiem el nom de les columnes
df.columns
```

```python
#Canviem la mesura superfície a km2
df['Superfície (ha)'] = df['Superfície (ha)']/100
#Reanomenem la variable
df = df.rename(columns = {'Superfície (ha)': 'Superficie (km2)'})

#Calculem les variables relatives i les noves. Les relatives les passem a percentatge (cas de població o emigració), o a equipaments per km2, com restaurants
df['Poblacio_sense_estudis'] = df['Poblacio_sense_estudis']/df['Poblacio_total']*100
df['Poblacio_univ'] = df['Poblacio_univ']/df['Poblacio_total']*100
df['Poblacio_20_40_univ_relatiu'] = df['Poblacio_20_40_univ']/df['Poblacio_20_40']*100
df['Poblacio_20_40_univ_global'] = df['Poblacio_20_40_univ']/df['Poblacio_total']*100
df['Poblacio_20_40'] = df['Poblacio_20_40']/df['Poblacio_total']*100
df['Total_arribats_BCN_inc_renda'] = df['Total_arribats_BCN_inc_renda']/df['Total_arribats_BCN']*100
df['Total_marxen'] = df['Total_marxen_barri_BCN'] + df['Total_marxen_fora_BCN']
df['Total_marxen_fora_BCN'] = df['Total_marxen_fora_BCN']/df['Total_marxen']*100
df['Total_expulsats'] = df['Total_expulsats']/df['Total_marxen']*100
df['Saldo_migratori'] = df['Total_arribats_BCN']-df['Total_marxen']
df['poblacio_immigrant'] = df['poblacio_immigrant']/df['Poblacio_total']*100
df['poblacio_IDH_sup'] = df['poblacio_IDH_sup']/df['Poblacio_total']*100
df['poblacio_IDH_inf'] = df['poblacio_IDH_inf']/df['Poblacio_total']*100
df['total_viuen_sols'] = df['total_viuen_sols']/df['Poblacio_total']*100
df['Allotjaments'] = df['Allotjaments']/df['Superficie (km2)']
df['Restaurants'] = df['Restaurants']/df['Superficie (km2)']
df['Taules_terrasses'] = df['Taules_terrasses']/df['Superficie (km2)']
df['Oci_nocturn'] = df['Oci_nocturn']/df['Superficie (km2)']
df['Teatres_espectacles'] = df['Teatres_espectacles']/df['Superficie (km2)']
df['Museus'] = df['Museus']/df['Superficie (km2)']
df['Parcs'] = df['Parcs']/df['Superficie (km2)']
df['Punts_interes_turistic'] = df['Punts_interes_turistic']/df['Superficie (km2)']
df['Comercos_quotidians'] = df['Comercos_quotidians']/df['Total_comercos']*100
df['Max_aforament'] = df['Max_aforament']/df['Superficie (km2)']
df['Avg_aforament'] = df['Avg_aforament']/df['Superficie (km2)']

#Reanomenem algunes de les variables per a què siguin més identificatives
df = df.rename(columns = {'Poblacio_20_40_univ_relatiu': 'Universitaris_20_40_relatiu',
                          'Poblacio_20_40_univ_global': 'Universitaris_20_40_total',
                          'Total_arribats_BCN_inc_renda': 'Immigrants_inc_renda', 
                          'Total_marxen': 'Emigrants',
                          'Total_marxen_fora_BCN': 'Emigrants_fora_BCN',
                          'Total_expulsats': 'Emigrants_Africa_Asia',
                          'total_viuen_sols': 'Empadronats_sols',
                          'Contractes': 'Contractes_lloguer'
                         }
              )

#Escollim i reordenem les variables que necessitem
columns = ['Codi_Barri', 'Nom_Barri', 'Any', 'Superficie (km2)', 'Poblacio_total', 'Poblacio_sense_estudis', 'Poblacio_univ',
           'Poblacio_20_40', 'Universitaris_20_40_relatiu', 'Universitaris_20_40_total', 'renda_mitjana', 'renda_Gini',
           'Saldo_migratori', 'poblacio_immigrant', 'Emigrants', 'Emigrants_fora_BCN', 'Emigrants_Africa_Asia', 'Immigrants_inc_renda',
           'poblacio_IDH_sup', 'poblacio_IDH_inf', 'Empadronats_sols', 'Contractes_lloguer', 'Preu_lloguer','Total_comercos', 
           'Comercos_quotidians', 'Restaurants', 'Taules_terrasses', 'Allotjaments', 'Oci_nocturn' ,'Teatres_espectacles',
           'Museus', 'Parcs', 'Punts_interes_turistic', 'Assistents_cultura', 'Espais_participacio', 'Max_soroll', 'Avg_soroll',
           'Max_aforament', 'Avg_aforament']

df = df[columns]
df.head()
```

```python
#Hem de convertir les columnes corresponents al valor d'indicador que cada filera contingui el valor d'un sol indicador per l'any corresponent
df = pd.melt(df, id_vars=['Codi_Barri', 'Nom_Barri', 'Any', 'Superficie (km2)'], var_name='Indicador', value_name='Valor')
df.head()
```

```python
#Exportem finalment el dataset
df.to_excel("../Data_cleaning/def_dataset/bcnGentTableau.xlsx", index=False)
```
